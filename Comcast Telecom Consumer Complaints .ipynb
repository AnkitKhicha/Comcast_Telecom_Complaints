#### Task 2 - Provide the trend chart for the number of complaints at monthly and daily granularity levels.

```python
# Comverting the date_month_year dtype from object to Dtype datetime64.MEans used to fomat the time stamp in string 
# format to pandas date time object.
cmctel_com['Date_month_year'] = pd.to_datetime(cmctel_com['Date_month_year'])
cmctel_com.info()
```

```python
## Extracting the month from the Date_month_year.
cmctel_com['Month'] = cmctel_com['Date_month_year'].dt.month
cmctel_com['Day'] = cmctel_com['Date_month_year'].dt.day
cmctel_com

#Extracting the month from the Date_month_year.
#cmctel_com['Month'] = cmctel_com.Date_month_year.dt.month
#cmctel_com['Day'] = cmctel_com.Date_month_year.dt.day
#cmctel_com
```

```python
## Extracting the month and Day from the Date_month_year can be done using function pd.DatetimeIndex also

cmctel_com['Month'] = pd.DatetimeIndex(cmctel_com['Date_month_year']).month
cmctel_com['Day'] = pd.DatetimeIndex(cmctel_com['Date_month_year']).day
cmctel_com['Year'] = pd.DatetimeIndex(cmctel_com['Date_month_year']).year
cmctel_com['Month_Name'] = cmctel_com.Date_month_year.dt.month.apply(lambda x : calendar.month_abbr[x])   
## Extracting the name(abbrevations) of the month by using lambda function.
cmctel_com
```

### Provide the trend chart for monthly and daily granulaties.

```python
## Extrating the monthly and daily no or counts of granulaties.value_counts() function will use for it to 
##count the no of complaints monthly.
Monthly_comp = cmctel_com['Month_Name'].value_counts().reset_index().rename({'index' : 'Months' ,'Month_Name' : 'MonthlyComplaints'},axis = 1)  
Monthly_comp.T

# We got complaints from month of max to month of min.
```

```python
Daily_comp = cmctel_com.groupby(['Month_Name','Day']).count()['City'].to_frame().reset_index().rename({'Month_Name' : 'Month' , 'Day' : 'Date' , 'City' : 'Perdaycomp'},axis = 1)
print(Daily_comp.T)
print(Daily_comp[Daily_comp.Month == 'Jan'])   # January month complaints
Daily_comp[Daily_comp.Month == 'Feb']          # February month complaints.

## Similarly we can find the other months daily complaints
```

#### Task5- Create a new categorical variable with value as Open and Closed. Open & Pending is to be categorized as Open and 
#### Closed & Solved is to be categorized as Closed.

```python
Comps_Status = cmctel_com['Status'].value_counts().reset_index().rename({'index' : 'Comp_Status' , 'Status' : 'Status_counts'},axis = 1)
Comps_Status
```

```python
## Creating a categorical variable for assigining the status with value open and close according to Complaint_Status.

cmctel_com['Final_Status'] = 0
for i in range(len(cmctel_com['Status'])):
    if(cmctel_com['Status'][i] == 'Closed')| (cmctel_com['Status'][i] == 'Solved'):
        cmctel_com['Final_Status'][i] = 'Closed'
    else:
        cmctel_com['Final_Status'][i] = 'Open'
cmctel_com[(cmctel_com.Final_Status == 'Closed') | (cmctel_com.Final_Status == 'Open')]  #Added a catog. variable 
```

### Task - 9 Provide the percentage of complaints resolved till date, which were received through the Internet and customer care calls.

```python
Comp_res_cuscare_net = cmctel_com.groupby(['Received Via', 'Final_Status']).count()['City'].to_frame().reset_index().rename({'City' : 'Comp_Counts'},axis = 1)
Comp_res_cuscare_net
```

```python
## Getting the unstacked Final Status of complaints.
Comp_via_cuscare_net = cmctel_com.groupby(['Received Via' , 'Final_Status']).size().unstack().fillna(0)
Comp_via_cuscare_net
```

#### Size () is the function in numpy package that returns the size of an array.The size of an array is the total number of elements 
#### in the array.

```python
# Percentage of resolved complaints received via Customer care calls and Internet.
Comp_via_cuscare_net['Resolved_complaints'] = Comp_via_cuscare_net['Closed']/Comp_via_cuscare_net['Closed'].sum()*100
Comp_via_cuscare_net
```

#### 50.61% Complaints resolved via Customet Care Call.
#### 49.38% Complaints resolved via Internet.

```python
Comp_via_cuscare_net.plot(kind = 'barh',stacked = True)
```

```python
# Pie Chart Visulisation.

fig , ax = plt.subplots(1 , 2,figsize = (10,30))

### Visulisation of complaints that is resolved via customer care calls and internet
ax[0].pie(Comp_via_cuscare_net['Resolved_complaints'].value_counts(),labels = Comp_via_cuscare_net.value_counts().index[:], autopct ='%1.1f%%')
plt.tight_layout()  
ax[0].set_title('Proportion of recived through Customer care call and Internet',weight='bold',color = 'darkorange',size = 13)

## Visulisation of complaints received through behalf of someone.
ax[1].pie(cmctel_com['Filing on Behalf of Someone'].value_counts() , labels = cmctel_com['Filing on Behalf of Someone'].value_counts().index[:],autopct = '%1.1f%%')
ax[1].set_title('Complaints Filing on Behalf of Someone',weight = 'bold',color = 'darkorange',size = 13)
plt.tight_layout()

```
